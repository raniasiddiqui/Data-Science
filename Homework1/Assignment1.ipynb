```python
import numpy as np
formatted_df = pd.read_csv('formatted_580SurveyCleanupFinal.csv')
formatted_df.isnull().sum()
# to identify and list all columns that have missing values.
```

```python
# Dropping columns that have empty values
formatted_df = formatted_df.drop(['Access Code', 'Email Address'], axis=1)
```

```python
# replacing values 
formatted_df['Q4'] = formatted_df['Q4'].replace({'7O':'70', '6l':'61', '5.2':'52'})
formatted_df.fillna({'Q4': formatted_df['Q4'].mode()[0]}, inplace=True)
formatted_df.head()
```

```python
formatted_df['Q4']
```

```python
# filling columns with missing values
columns_to_fill = ['Q8', 'Q9', 'Q10', 'Q11', 'Q12']

for column in columns_to_fill:
    formatted_df.fillna({column:formatted_df[column].mode()[0]}, inplace=True)
formatted_df.head()
```

```python
columns_to_fill = ['Region', 'City / Town / District']

for column in columns_to_fill:
    formatted_df.fillna({column:formatted_df[column].mode()[0]}, inplace=True)
formatted_df.head()
```

```python
formatted_df['Q3'] = formatted_df['Q3'].str.replace(' ', '') #filling spaces with Q3
formatted_df['Q3'] = formatted_df['Q3'].replace('1993', '28') #replace 1993 with 28
formatted_df.fillna({'Q3': formatted_df['Q3'].mode()[0]}, inplace=True) #filling missing values with mode
formatted_df.head()
```

```python
formatted_df['Q6'] = formatted_df['Q6'].str.title().str.strip().replace({
    'Bachelor of engineering ': 'Engineering',
    'Engineering ': 'Engineering',
    'Cse': 'Computer Science and Engineering',
    'Computer science engineering ': 'Computer Science and Engineering',
    'Computer engineering': 'Computer Science and Engineering',
    'B.Tech In Cs': 'Computer Science',
    'Criminology law and soci': 'Criminology law and social sciences',
    'Ece': 'Electrical and Computer Engineering',
    'Economic': 'Economics',
    'Bit': 'Business in Technology',
    'Yes': np.NaN,
})

formatted_df.fillna({'Q6': formatted_df['Q6'].mode()[0]}, inplace=True)
formatted_df.head()
```

```python
formatted_df.fillna({'Q7': 'Not responded'}, inplace=True)
formatted_df['Q8'] = formatted_df['Q8'].replace('Microsoft/Windows', 'Windows')
formatted_df['Q8'] = formatted_df['Q8'].replace('Apple/MacBook', 'Macbook')
formatted_df['Q8'] = formatted_df['Q8'].replace('Acer/Windows', 'Windows')

formatted_df.fillna({'Q15': 'Not responded'}, inplace=True)

formatted_df.head()
```

```python
formatted_df['Q7']
```

```python
formatted_df['Q5'] = formatted_df['Q5'].str.capitalize()
formatted_df['Q5'] = formatted_df['Q5'].fillna('N/A')
formatted_df['Q5'] = formatted_df['Q5'].replace(['Usa', 'United states', 'United states ', 'Us'], 'USA')
formatted_df['Q5'] = formatted_df['Q5'].replace(['United states ; pakistan '], 'USA ; Pakistan')
```

```python
formatted_df['Q5']
```

```python
#Making CSV file

formatted_df.to_csv('cleaned_580SurveyCleanupFinal.csv', index = False)
```
